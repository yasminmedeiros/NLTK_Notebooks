```python
def ConvertTexttoPigLatin (text):
    arr = text.split(' ')
    for x in range(len(arr)):
        y = []
        for z in range(len(arr[x])):
            if z!=0:
                y.append(arr[x][z])
        y.append(arr[x][0])
        y.append('ay')
        arr[x] = ''.join(y)
    print(' '.join(arr))
```

```python
ConvertTexttoPigLatin('Oi Yasmin Tudo bom com vocÊ')
```

```python
def indicePorPalavraDecadaFrase (arrFrases):
    us = sum([len(x) for x in arrFrases])
    for palavras in arrFrases:
        uw = sum([len(x) for x in palavras])
        ari = 4.71*uw +0.5*us -21.43
        print('ari = {}\n'.format(ari))
indicePorPalavraDecadaFrase(nltk.corpus.brown.sents ())        
        
```
