```python
for fileid in gutenberg.fileids():
    num_chars = len(gutenberg.raw(fileid)) #nos diz quantas letras ocorrem no texto, incluindo os espaços entre as palavras
    num_words = len(gutenberg.words(fileid))# número de palavras
    num_sents = len(gutenberg.sents(fileid))#divide o texto em suas frases, onde cada frase é uma lista de palavras
    num_vocab = len(set(w.lower() for w in gutenberg.words(fileid)))
    print(round (num_chars / num_words), round (num_words / num_sents)
      , round (num_words / num_vocab), fileid)    
# comprimento médio das palavras, comprimento médio das frases e o número de vezes que cada item de vocabulário aparece no texto em média (nossa pontuação de diversidade lexical)
```

```python
news_text = brown.words(categories = 'news')
fdist = nltk.FreqDist(w.lower() for w in news_text)
modals = ['can', 'could', 'may', 'might', 'must', 'will']
for m in modals:
    print(m + ':',fdist[m], end = ' ')# esse end é pra colocar tudo em uma única linha
```

```python
cfd = nltk.ConditionalFreqDist(
          (genre, word)
           for genre in brown.categories()
           for word in brown.words(categories=genre))
genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
cfd.tabulate(conditions=genres, samples=modals)
```

Quando os textos de um corpus são divididos em várias categorias, por gênero, tópico, autor, etc., podemos manter distribuições de frequência separadas para cada categoria. Isso nos permitirá estudar diferenças sistemáticas entre as categorias. Na seção anterior, conseguimos isso usando o tipo de dados ConditionalFreqDist do NLTK . Uma distribuição de frequência condicional é uma coleção de distribuições de frequência, cada uma para uma "condição" diferente. A condição geralmente será a categoria do texto. 2.1 representa um fragmento de uma distribuição de frequência condicional com apenas duas condições, uma para texto de notícias e outra para texto de romance.

```python
#ConditionalFreqDist () pega uma lista de pares
from nltk.corpus import brown
cfd = nltk.ConditionalFreqDist(
          (genre, word)
           for genre in brown.categories()
           for word in brown.words(categories=genre))
```

Você deve ter notado que as expressões de várias linhas que usamos com distribuições de frequência condicionais parecem compreensões de lista, mas sem os colchetes. Em geral, quando usamos a compreensão da lista como parâmetro para uma função, como set ([w.lower () para w em t])) , podemos omitir os colchetes e escrever: set (w.lower ( ) para w em t) 

A função bigrams () pega uma lista de palavras e cria uma lista de pares de palavras consecutivos. Lembre-se de que, para ver o resultado e não um "objeto gerador" enigmático, precisamos usar a função list ()

```python
def generate_model(cfdist, word, num=15):
    for i in range(num):
        print(word, end = ' ')
        word = cfdist[word].max()
text = nltk.corpus.genesis.words('english-kjv.txt')
bigrams = nltk.bigrams(text)
cfd = nltk.ConditionalFreqDist(bigrams)
print(cfd['living'])
generate_model(cfd,'living')
```

 Quando chamamos a função, escolhemos uma palavra (como 'viva' ) como nosso contexto inicial e, uma vez dentro do loop, imprimimos o valor atual da palavra variável e redefinimos a palavra para ser o token mais provável nesse contexto. (usando max ()); da próxima vez, usaremos essa palavra como nosso novo contexto. Como você pode ver ao inspecionar a saída, essa abordagem simples da geração de texto tende a ficar presa nos loops; outro método seria escolher aleatoriamente a próxima palavra dentre as palavras disponíveis.

```python
def lexical_diversity(text):
    return len(text)/len(set(text))
```

```python
def lexical_diversity(my_text_data):
    word_count = len(my_text_data)
    vocab_size = len(set(my_text_data))
    diversity_score = vocab_size/word_count
    return diversity_score
```

```python
from nltk.corpus import genesis
kvj = genesis.words('english-kjv.txt')
lexical_diversity(kvj)
```

```python
def plural(word):
    if word.endswith('y'):
        return word[:-1] + 'ies'
    elif word[-1] in 'sx' or word[-2:] in ['sh', 'ch']:
        return word + 'es'
    elif word.endswith('an'):
        return word[:-2] + 'en'
    else:
        return word + 's'
#tem erros ainda
```

Filtrando um texto: este programa calcula o vocabulário de um texto e remove todos os itens que ocorrem em uma lista de palavras existente, deixando apenas as palavras incomuns ou incorretas.

```python
def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in nltk.corpus.words.words())
    unusual = text_vocab - english_vocab
    return sorted(unusual)
```

```python
unusual_words(nltk.corpus.gutenberg.words('austen-sense.txt'))
```

```python
unusual_words(nltk.corpus.nps_chat.words())
```

```python
def content_fraction(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords ]
    return len(content)/len(text)
content_fraction(nltk.corpus.reuters.words())
```
