Função	Significado<br><br>
s.inicia com (t)	teste se s começa com t<br>
s.endswith (t)	teste se s termina com t
t em s	teste se t é uma subcadeia de s<br>
s.islower ()	teste se s contém caracteres em maiúsculas e minúsculas<br>
s.isupper ()	testar se s contém caracteres em maiúsculas e minúsculas<br>
s.isalpha ()	teste se s está vazio e todos os caracteres em s são alfabéticos<br>
s.isalnum ()	teste se s está vazio e todos os caracteres em s são alfanuméricos<br>
s.isdigit ()	teste se s está vazio e todos os caracteres em s são dígitos<br>
s.istitle ()	testar se s contém caracteres em maiúsculas e minúsculas (ou seja, todas as palavras em s têm maiúsculas iniciais)<br>
