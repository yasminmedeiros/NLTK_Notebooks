```python
from nltk.corpus import inaugural
cfd = nltk.ConditionalFreqDist(
           (target, fileid[:4])
           for fileid in inaugural.fileids()
           for w in inaugural.words(fileid)
           for target in ['men','woman','people']
           if w.lower().startswith(target))
cfd.plot()
```

```python
def tokens_types_lexicalDiversity(genre):
    types = len(set(brown.words(categories=genre)))
    tokens = len(brown.words(categories=genre))
    lexicalDiversity = types/tokens
    return types,tokens, lexicalDiversity
```

```python
types = {}
tokens = {}
lexicalDiversity = {}
for genre in brown.categories():
    ty,to, lexical = tokens_types_lexicalDiversity(genre)
    types[genre] = ty
    tokens[genre] = to
    lexicalDiversity[genre] = lexical
print(types,tokens,lexicalDiversity)
```

```python
import pandas as pd
pd.DataFrame([tokens,types, lexicalDiversity])
```

```python
modals = ['can', 'could', 'may', 'might', 'must', 'will']
for genre in brown.categories():
    fd = nltk.FreqDist([a for a in
                        brown.words(categories=genre) 
                        if a in  modals])
    print('\n'+ genre)
    fd.tabulate()
```
